## Read Data and Cleaning Data

```python
# how to read data in google sheet

df = pd.read_csv("Games Sales - Case Study - Games (1) (1).csv")
df.head(20)
```

```python
df.info
```

```python
df.columns
```

```python
# change type date

df_cleaned = df.copy()

df_cleaned['Release'] = pd.to_datetime(df_cleaned['Release'])

print(df_cleaned.info())
df_cleaned.head(10)
```

```python
#typo check

df_cleaned['Series'].value_counts()
```

```python
for col_name in ['Name', 'Sales', 'Series', 'Genre', 'Developer',
       'Publisher']:
       print(df_cleaned[col_name].value_counts(),'\n')
```

```python
#replace

dict_typo = {'Real-time strategy' : 'RTS',
             'Action role-playing game' : 'ARPG',
             'Role-playing game' : 'RPG',
             'First-person shooter' : 'FPS',
             'Construction and management simulation' : 'CMS',
             'Action role-playing' : 'ARPG',
             'Space combat simulation' : 'Space flight simulation',
             'Maze, arcade' : 'Maze arcade',
             'Action-adventure, platformer' : 'Action-adventure',
             'Action-adventure, Survival' : 'Action-adventure'}

df_replaced = df_cleaned.replace(dict_typo)
df_replaced.head(30)
```

```python
df_replaced['Genre'].value_counts()
```

#EDA

## oldest and newest games

```python
# Find the oldest and newest games

oldest_game_index = df_cleaned['Release'].idxmin()
newest_game_index = df_cleaned['Release'].idxmax()

oldest_game = df.loc[oldest_game_index]
newest_game = df.loc[newest_game_index]

print('Oldest game info:')
print(oldest_game)
print('Newest game info:')
print(newest_game)
```

## most published publisher

```python
most_published_publisher = df['Publisher'].value_counts().idxmax()
print('Most published publisher is ' + most_published_publisher)
```

## most developed developer

```python
most_developed_developer = df['Developer'].value_counts().idxmax()
print('Most developed developer is ' + most_developed_developer)
```

## most selling series

```python
series_most_sales = df.groupby('Series')['Sales'].sum().idxmax()
print(series_most_sales + ' is the most-selling series')
```

## Most selling game

```python
series_most_games = df['Series'].value_counts().idxmax()
print(series_most_games + ' is the most-selling game')
```
